## importing libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Load dataset

from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer.keys()
```

```python
cancer['data']
```

```python
cancer['feature_names']
```

```python
print(cancer['DESCR'])
```

```python
df_feature = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df_feature
```

```python
cancer['target_names']
```

```python
df_target = pd.DataFrame(cancer['target'], columns=['is_malignant_cancer'])
df_target
```

```python
## Exploring the dataset
df_feature.info()
```

```python
from sklearn.model_selection import train_test_split

X = df_feature
y = df_target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
## Evaluation

from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

## Run VSM with "linear", C="10"


```python
from sklearn.metrics import accuracy_score
linear_svc = SVC(kernel='linear', C=10.0)
linear_svc.fit(X_train, y_train)
y_pred_svc = linear_svc.predict(X_test)

print("Linear model accuracy:", accuracy_score(y_test, y_pred_svc))
```

```python
linear_svc.score(X_train, y_train)
```

```python
 
linear_svc = SVC(kernel='rbf', C=11.0)
linear_svc.fit(X_train, y_train)
y_pred_svc = linear_svc.predict(X_test)

print("Linear model accuracy:", accuracy_score(y_test, y_pred_svc))
```

```python
 
linear_svc = SVC(kernel='poly', C=13.0)
linear_svc.fit(X_train, y_train)
y_pred_svc = linear_svc.predict(X_test)

print("Linear model accuracy:", accuracy_score(y_test, y_pred_svc))
```

```python
linear_svc = SVC(kernel='sigmoid', C=13.0)
linear_svc.fit(X_train, y_train)
y_pred_svc = linear_svc.predict(X_test)

print("Linear model accuracy:", accuracy_score(y_test, y_pred_svc))
```

```python
from sklearn.model_selection import GridSearchCV
para_grid= {
    'C': [1,2,4,6,8,10],
    'kernel': ['rbf', 'linear', 'poly', 'sigmoid'],
}

opt_model = GridSearchCV(SVC(), para_grid)

opt_model.fit(X_train, y_train)

y_pred_opt = opt_model.predict(X_test)
```

```python
opt_model.best_params_
```

```python
opt_model.score(X_train, y_train)
```

```python
query = df_feature.loc[100:100]
query
```

```python
result = opt_model.predict(query)

if result ==1:
    print("Malignant Cancer Detected")
else:
    print("Benign Cancer Detected")
```
