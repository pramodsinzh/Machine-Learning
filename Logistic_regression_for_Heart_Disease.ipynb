```python
# Convert target labels to binary values
heartDisease["Heart Disease"] = heartDisease["Heart Disease"].map({"Presence": 1, "Absence": 0})
heartDisease.head()
```

```python
categorical_cols = [
    "Sex",
    "Chest pain type",
    "FBS over 120",
    "EKG results",
    "Exercise angina",
    "Slope of ST",
    "Number of vessels fluro",
    "Thallium",
]
heartDisease_model = pd.get_dummies(heartDisease, columns=categorical_cols, drop_first=True)
heartDisease_model.head()
```

```python
X = heartDisease_model.drop("Heart Disease", axis=1)
y = heartDisease_model["Heart Disease"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=1000)
```

```python
log_model.fit(X_train, y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score

print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.3f}")
```
